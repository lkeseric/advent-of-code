```python
import ast
import copy
import re

from collections import defaultdict
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from aocd import get_data, submit

DAY = 3
YEAR = 2023
```

```python
# use test data
raw_test = """input"""

# use real data
raw = get_data(day=DAY, year=YEAR)

print(raw_test)
```

```python
def parse_data(data):
    # data = data.split("\n")
    # data = [d.split(" ") for d in data]
    # data = [d.split('\n') for d in data.split('\n\n')]
    # data = [np.array(d.split('\n'), dtype=int) for d in data.split('\n\n')]

    return data


dummy = parse_data(raw_test)
real = parse_data(raw)

real
```

# Part 1

```python
matrix_data = real.split("\n")
number_sum = 0
search_window = []
part_number = False 
number_string = ""

#find number
for i, row in enumerate(matrix_data):
    
    for j, item in enumerate(row):
        # get number
        if item.isdigit():
            if number_string == "":
                search_window.append(max(0, j-1))
            if j == len(row)-1:
                search_window.append(j+1)
            number_string += item
        # after we get the whole number number
        elif number_string != "":
            search_window.append(j+1)
            
        if len(search_window) == 2:
            #check top
            if i != 0:
                for symbol in matrix_data[i-1][search_window[0]:search_window[1]]:
                    if not(symbol.isdigit() or symbol == '.'):
                        part_number = True
                        break
            #check sides
            for symbol in matrix_data[i][search_window[0]:search_window[1]]:
                if not(symbol.isdigit() or symbol == '.'):
                    part_number = True
                    break
            #check bottom
            if i != len(matrix_data)-1:
                for symbol in matrix_data[i+1][search_window[0]:search_window[1]]:
                    if not(symbol.isdigit() or symbol == '.'):
                        part_number = True
                        break
            #if true add to sum
            if part_number:
                number_sum += int(number_string)
                #print(int(number_string))
                part_number = False
            number_string = ""
            search_window = []

print(number_sum)
result = number_sum
```

```python
submit(result, part="a", day=DAY, year=YEAR)
```

# Part 2

```python
#reuse code from before
matrix_data = real.split("\n")
ratios_sum = 0
search_window = []
part_number = False 
number_string = ""
gear_loc = defaultdict(int)

#find number
for i, row in enumerate(matrix_data):
    
    for j, item in enumerate(row):
        # get number
        if item.isdigit():
            if number_string == "":
                search_window.append(max(0, j-1))
            if j == len(row)-1:
                search_window.append(j+1)
            number_string += item
        # after we get the whole number number
        elif number_string != "":
            search_window.append(j+1)
            
        if len(search_window) == 2:
            #check top
            if i != 0:
                for k, symbol in enumerate(matrix_data[i-1][search_window[0]:search_window[1]]):
                    if symbol == '*':
                        if gear_loc[i-1, search_window[0]+k] == 0:
                            gear_loc[i-1, search_window[0]+k] += int(number_string)
                        else:
                            gear_loc[i-1, search_window[0]+k] *= int(number_string)
                            ratios_sum += gear_loc[i-1, search_window[0]+k]
            #check sides
            for k, symbol in enumerate(matrix_data[i][search_window[0]:search_window[1]]):
                if symbol == '*':
                    if gear_loc[i, search_window[0]+k] == 0:
                        gear_loc[i, search_window[0]+k] += int(number_string)
                    else:
                        gear_loc[i, search_window[0]+k] *= int(number_string)
                        ratios_sum += gear_loc[i, search_window[0]+k]
            #check bottom
            if i != len(matrix_data)-1:
                for k, symbol in enumerate(matrix_data[i+1][search_window[0]:search_window[1]]):
                    if symbol == '*':
                        if gear_loc[i+1, search_window[0]+k] == 0:
                            gear_loc[i+1, search_window[0]+k] += int(number_string)
                        else:
                            gear_loc[i+1, search_window[0]+k] *= int(number_string)
                            ratios_sum += gear_loc[i+1, search_window[0]+k]
            number_string = ""
            search_window = []

print(ratios_sum)
result = ratios_sum
```

```python
submit(result, part="b", day=DAY, year=YEAR)
```
