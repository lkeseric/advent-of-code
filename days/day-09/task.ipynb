```python
import ast
import copy
import re

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from aocd import get_data, submit

DAY = 9
YEAR = 2023
```

```python
# use test data
raw_test = """input"""

# use real data
raw = get_data(day=DAY, year=YEAR)

print(raw)
```

```python
def parse_data(data):
    data = data.split("\n")
    data = [d.split(" ") for d in data]
    # data = [d.split('\n') for d in data.split('\n\n')]
    # data = [np.array(d.split('\n'), dtype=int) for d in data.split('\n\n')]

    return data


dummy = parse_data(raw_test)
real = parse_data(raw)

real
```

# Part 1

```python
def calc_diff(num_list):
    diff_list = []
    old_num = num_list[0]
    for i in range(1, len(num_list)):
        new_num = num_list[i]
        diff = new_num - old_num
        diff_list.append(diff)
        
        old_num = new_num
        
    return diff_list

sum_of_next_numbers = 0
for row in real:
    number_list = [int(i) for i in row]
    last_number_list = [number_list[-1]]
    while (not(all(np.array(number_list) == 0))):
        number_list = calc_diff(number_list)
        last_number_list.append(number_list[-1])
    
    next_number = np.sum(last_number_list)
    sum_of_next_numbers += next_number

result = int(sum_of_next_numbers)
print(result)
```

```python
submit(result, part="a", day=DAY, year=YEAR)
```

# Part 2

```python
sum_of_prev_numbers = 0
for row in real:
    number_list = [int(i) for i in row]
    prev_number_list = [number_list[0]]
    while (not(all(np.array(number_list) == 0))):
        number_list = calc_diff(number_list)
        prev_number_list.append(number_list[0])
        #print(number_list)
    
    #print(prev_number_list)
    prev_number = prev_number_list.pop()
    while prev_number_list:
        prev_number = prev_number_list.pop() - prev_number
        #print(prev_number)
    sum_of_prev_numbers += prev_number

result = int(sum_of_prev_numbers)
print(result)
```

```python
submit(result, part="b", day=DAY, year=YEAR)
```
