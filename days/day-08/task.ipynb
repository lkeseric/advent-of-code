```python
import ast
import copy
import re

import math
from itertools import cycle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from aocd import get_data, submit

DAY = 8
YEAR = 2023
```

```python
# use test data
raw_test = """input"""

# use real data
raw = get_data(day=DAY, year=YEAR)

print(raw_test)
```

```python
def parse_data(data):
    #data = data.split("\n")
    # data = [d.split(" ") for d in data]
    data = [d.split('\n') for d in data.split('\n\n')]
    # data = [np.array(d.split('\n'), dtype=int) for d in data.split('\n\n')]

    return data


dummy = parse_data(raw_test)
real = parse_data(raw)

real
```

# Part 1

```python
instructions = ""
nodes = {}

for i, row in enumerate(real):
    if i == 0:
        instructions = real[0]
    if i >= 2:
        parent = row.split(" ")[0]
        left_child = row.split(" ")[2][1:4]
        right_child = row.split(" ")[3][:3]
        nodes[parent] = (left_child, right_child)
```

```python
step_counter = 0
instructions_iter = cycle(instructions)
parent = 'AAA'


while (parent != "ZZZ"):
    
    step_counter += 1
    (left, right) = nodes[parent]
    step = next(instructions_iter)

    if step == "L":
        parent = left
    elif step == "R":
        parent = right
        
result = step_counter
print(parent, result)
```

```python
submit(result, part="a", day=DAY, year=YEAR)
```

# Part 2

```python
def get_start_nodes(nodes):
    starting_nodes = []
    for name in sorted(list(nodes.keys())):
        if name[2] == "A":
            starting_nodes.append(name)
            
    return starting_nodes
        
def end_state(nodes):
    for node in nodes:
        if node[2] == "Z":
            pass
        else:
            return False
    
    return True

def end_state_partial(node):
    if node[2] == "Z":
        return True
    return False

step_counter = 0
instructions_iter = cycle(instructions)
parent_nodes = get_start_nodes(nodes)
lcm = np.zeros(len(parent_nodes))

while (not(all(lcm > 0))):
    
    step = next(instructions_iter)
    binary_nodes = ""
    
    for i, parent in enumerate(parent_nodes):
        if end_state_partial(parent):
            binary_nodes += "1"
        else:
            binary_nodes += "0"
        (left, right) = nodes[parent]
        if step == "L":
            parent_nodes[i] = left
        elif step == "R":
            parent_nodes[i] = right

    if binary_nodes != "000000":
        b = binary_nodes.index("1")        
        if lcm[b] == 0:
            lcm[b] += step_counter
            
    step_counter += 1
    
        
result = int(np.lcm.reduce(lcm.astype(np.int64)))
print(result)
```

```python
submit(result, part="b", day=DAY, year=YEAR)
```
