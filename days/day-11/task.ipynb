```python
import ast
import copy
import re

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from aocd import get_data, submit

DAY = 11
YEAR = 2023
```

```python
# use test data
raw_test = """input"""

# use real data
raw = get_data(day=DAY, year=YEAR)

print(raw_test)
```

```python
def parse_data(data):
    data = data.split("\n")
    # data = [d.split(" ") for d in data]
    # data = [d.split('\n') for d in data.split('\n\n')]
    # data = [np.array(d.split('\n'), dtype=int) for d in data.split('\n\n')]

    return data


dummy = parse_data(raw_test)
real = parse_data(raw)

real
```

# Part 1

```python
#140x140 matrix

# fill matrix with chars
universe = np.array([ list(row) for row in real ])

#find empty rows and columns
empty_rows = []
empty_columns = []
for i in range(len(universe)):
    if all(universe[:,i] == '.'): #rows
        empty_rows.append(i)
    if all(universe[i,:] == '.'): #columns
        empty_columns.append(i)

#expand universe
expanded_universe = universe.copy()

for i, j in enumerate(empty_columns):
    shift = i+j
    expanded_universe = np.concatenate((expanded_universe[:shift,:], expanded_universe[shift,:].reshape(1,-1), expanded_universe[shift:,:]), axis = 0)

for i,j in enumerate(empty_rows):
    shift = i+j
    expanded_universe = np.concatenate((expanded_universe[:,:shift], expanded_universe[:,shift].reshape(-1,1), expanded_universe[:,shift:]), axis = 1)

# mark galaxies
galaxies = {}
count = 0
for i in range(expanded_universe.shape[0]):
    for j in range(expanded_universe.shape[1]):
        if expanded_universe[i][j] == "#":
            count += 1
            galaxies[count] = (i, j)

# calculate pairwise shortest manhattan distance between galaxies (and sum it)
def manhattan_distance(coordinates1, coordinates2):
    (x1, y1) = coordinates1
    (x2, y2) = coordinates2
    return abs(x2-x1)+abs(y2-y1)

sum_of_shortest_paths = 0
for galaxy_coordinates1 in galaxies.values():
    for galaxy_coordinates2 in galaxies.values():
        sum_of_shortest_paths += manhattan_distance(galaxy_coordinates1, galaxy_coordinates2)

result = int(sum_of_shortest_paths/2) #divide by 2 since we counted every path twice
print(result)
```

```python
submit(result, part="a", day=DAY, year=YEAR)
```

# Part 2

```python
expanded_space = "0"

# fill matrix with chars
universe = np.array([ list(row) for row in real ])

#find empty rows and columns
empty_rows = []
empty_columns = []
for i in range(len(universe)):
    if all(universe[:,i] == '.'): #rows
        empty_rows.append(i)
    if all(universe[i,:] == '.'): #columns
        empty_columns.append(i)

#expand universe
expanded_universe = universe.copy()
for i, j in enumerate(empty_columns):
    shift = i+j
    expanded_universe = np.concatenate((expanded_universe[:shift,:], np.full_like(expanded_universe[shift,:], expanded_space).reshape(1,-1), expanded_universe[shift:,:]), axis = 0)

for i,j in enumerate(empty_rows):
    shift = i+j
    expanded_universe = np.concatenate((expanded_universe[:,:shift], np.full_like(expanded_universe[:,shift], expanded_space).reshape(-1,1), expanded_universe[:,shift:]), axis = 1)

# mark galaxies
galaxies = {}
count = 0
for i in range(expanded_universe.shape[0]):
    for j in range(expanded_universe.shape[1]):
        if expanded_universe[i][j] == "#":
            count += 1
            galaxies[count] = (i, j)


# calculate distance between galaxies
def increment_value(value1, value2):
    if value1 > value2:
        value1 -= 1
    else:
        value1 += 1
    return value1

def custom_distance_calculator(coordinates1, coordinates2):
    expansion_factor = 1000000
    (x1, y1) = coordinates1
    (x2, y2) = coordinates2
    
    #higher_x, lower_x = sort_values(x1, x2)
    #higher_y, lower_y = sort_values(y1, y2)
    
    distance_x = 0
    distance_y = 0
    
    while ((x1 != x2) or (y1 != y2)):
        if x1 != x2:
            x1 = increment_value(x1,x2) 
            if expanded_universe[x1][y1] == expanded_space:
                distance_x += expansion_factor
                x1 = increment_value(x1,x2) 
                # we are jumping over the expanded space and one more space
                # this is for easier counting
            else:
                distance_x += 1
            
            
        if y1 != y2:
            y1 = increment_value(y1,y2) 
            if expanded_universe[x1][y1] == expanded_space:
                distance_y += expansion_factor
                y1 = increment_value(y1,y2) 
            else:
                distance_y += 1
            
            
    return distance_x+distance_y
    
    
sum_of_shortest_paths = 0
for galaxy_coordinates1 in galaxies.values():
    for galaxy_coordinates2 in galaxies.values():
        sum_of_shortest_paths += custom_distance_calculator(galaxy_coordinates1, galaxy_coordinates2)
        
result = int(sum_of_shortest_paths/2)    
print(result)    
```

```python
submit(result, part="b", day=DAY, year=YEAR)
```
